```python
# ============================================================
# (Google Colab) CSV ì—…ë¡œë“œ -> Mistral ìš”ì•½(2ì¤„)+í‚¤ì›Œë“œ -> CSV ì €ì¥/ë‹¤ìš´ë¡œë“œ
# Colab ê¸°ë³¸ ê²½ë¡œ: /content
# ============================================================

# 1) ì„¤ì¹˜
!pip -q install pandas tqdm mistralai pydantic

# 2) CSV ì—…ë¡œë“œ
from google.colab import files
uploaded = files.upload()

if not uploaded:
    raise ValueError("âŒ CSV íŒŒì¼ì´ ì—…ë¡œë“œë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤. íŒŒì¼ì„ ì„ íƒí•´ ì—…ë¡œë“œí•˜ì„¸ìš”.")

filename = list(uploaded.keys())[0]
INPUT_PATH = f"/content/{filename}"   # âœ… Colab ê¸°ë³¸ ì—…ë¡œë“œ ê²½ë¡œ
print("âœ… INPUT_PATH =", INPUT_PATH)

# ì—…ë¡œë“œ íŒŒì¼ í™•ì¸
import os
print("ğŸ“ /content files:", os.listdir("/content"))


# 3) API Key í™•ì¸ (ë¯¸ë¦¬ %env ë˜ëŠ” Secretsë¡œ ì„¤ì •ë˜ì–´ ìˆì–´ì•¼ í•¨)
import re, json, time
import pandas as pd
from tqdm import tqdm
from pydantic import BaseModel, Field
from mistralai import Mistral

api_key = os.getenv("MISTRAL_API_KEY", "Y2nJU0YqjbedsHpedWe76Y4PPGtllniO").strip()
if not api_key:
    raise ValueError(
        "âŒ MISTRAL_API_KEYê°€ ë¹„ì–´ìˆìŠµë‹ˆë‹¤.\n"
        "ë¨¼ì € ì•„ë˜ì²˜ëŸ¼ ì‹¤í–‰í•˜ì„¸ìš”:\n"
        "%env MISTRAL_API_KEY=sk-....\n"
        "ë˜ëŠ” Colab Secrets(ğŸ”‘)ì— MISTRAL_API_KEYë¥¼ ì €ì¥í•˜ì„¸ìš”."
    )

client = Mistral(api_key=api_key)


# 4) CSV ë¡œë“œ + ì»¬ëŸ¼ ìë™ íƒì§€
df = pd.read_csv(INPUT_PATH)

CONTENT_COL_CANDIDATES = ["content", "body", "text", "ë³¸ë¬¸", "ë‚´ìš©"]
TITLE_COL_CANDIDATES   = ["title", "ì œëª©"]
URL_COL_CANDIDATES     = ["url", "link", "URL"]

def pick_col(cols, cands):
    for c in cands:
        if c in cols:
            return c
    return None

content_col = pick_col(df.columns, CONTENT_COL_CANDIDATES)
title_col   = pick_col(df.columns, TITLE_COL_CANDIDATES)
url_col     = pick_col(df.columns, URL_COL_CANDIDATES)

if content_col is None:
    raise ValueError(
        f"âŒ ë³¸ë¬¸ ì»¬ëŸ¼ì„ ì°¾ì§€ ëª»í–ˆìŠµë‹ˆë‹¤.\n"
        f"í˜„ì¬ ì»¬ëŸ¼: {list(df.columns)}\n"
        f"ë³¸ë¬¸ í›„ë³´: {CONTENT_COL_CANDIDATES}"
    )

print("âœ… Detected:", {"content_col": content_col, "title_col": title_col, "url_col": url_col})


# 5) ì„¤ì •ê°’
MODEL = "ministral-8b-latest"
MAX_CHARS = 8000
MAX_RETRY = 4
SLEEP_BETWEEN_CALLS = 0.2

OUTPUT_PATH = "/content/korea_policy_articles_summarized.csv"


# 6) ì „ì²˜ë¦¬
def norm(x):
    if x is None:
        return ""
    try:
        if pd.isna(x):
            return ""
    except Exception:
        pass
    x = str(x)
    x = re.sub(r"\s+", " ", x).strip()
    return x


# 7) êµ¬ì¡°í™” ì¶œë ¥ ìŠ¤í‚¤ë§ˆ
class SummaryOut(BaseModel):
    summary_line1: str = Field(..., description="í•œêµ­ì–´ ìš”ì•½ 1ì¤„(120ì ë‚´ì™¸)")
    summary_line2: str = Field(..., description="í•œêµ­ì–´ ìš”ì•½ 2ì¤„(120ì ë‚´ì™¸)")
    keywords: list[str] = Field(..., description="í‚¤ì›Œë“œ 5~8ê°œ, ì¤‘ë³µ ì—†ì´ ëª…ì‚¬/ëª…ì‚¬êµ¬")

SYSTEM_PROMPT = """ë„ˆëŠ” ì •ë¶€ ì •ì±… ë‰´ìŠ¤ì˜ í•µì‹¬ì„ ë¹ ë¥´ê²Œ ì •ë¦¬í•˜ëŠ” ë¶„ì„ê°€ë‹¤.
ë°˜ë“œì‹œ í•œêµ­ì–´ë¡œë§Œ ë‹µí•œë‹¤.
ìš”êµ¬ì‚¬í•­:
- summary_line1, summary_line2: ê° 1ì¤„ì”©, ê°ê° 120ì ë‚´ì™¸.
- keywords: 5~8ê°œ, ì¤‘ë³µ ì—†ì´, ëª…ì‚¬/ëª…ì‚¬êµ¬ ì¤‘ì‹¬.
- ì›ë¬¸ì— ì—†ëŠ” ì‚¬ì‹¤ì„ ì¶”ì •/ì°½ì‘í•˜ì§€ ë§ˆë¼.
"""

def mistral_one(title, url, content):
    user_msg = f"""[ì œëª©]
{title}

[URL]
{url}

[ë³¸ë¬¸]
{content}
"""
    resp = client.chat.parse(
        model=MODEL,
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": user_msg},
        ],
        response_format=SummaryOut,
        temperature=0,
        max_tokens=256,
    )

    try:
        return resp.choices[0].message.parsed
    except Exception:
        raw = resp.choices[0].message.content
        if isinstance(raw, str):
            data = json.loads(raw)
            return SummaryOut(**data)
        raise


# 8) ê²°ê³¼ ì»¬ëŸ¼ ì¤€ë¹„(ì´ì–´í•˜ê¸° ê°€ëŠ¥)
for c in ["summary_line1", "summary_line2", "keywords", "mistral_error"]:
    if c not in df.columns:
        df[c] = ""

def done(row):
    return bool(norm(row.get("summary_line1"))) and bool(norm(row.get("summary_line2")))


# 9) ì „ì²´ ì²˜ë¦¬
for i, row in tqdm(df.iterrows(), total=len(df)):
    if done(row):
        continue

    title = norm(row[title_col]) if title_col else ""
    url   = norm(row[url_col])   if url_col else ""
    text  = norm(row[content_col])

    if not text:
        df.at[i, "mistral_error"] = "EMPTY_CONTENT"
        continue

    text = text[:MAX_CHARS]

    last_err = None
    for attempt in range(1, MAX_RETRY + 1):
        try:
            out = mistral_one(title, url, text)
            df.at[i, "summary_line1"] = norm(out.summary_line1)
            df.at[i, "summary_line2"] = norm(out.summary_line2)
            df.at[i, "keywords"]      = ", ".join([norm(k) for k in out.keywords if norm(k)])
            df.at[i, "mistral_error"] = ""
            last_err = None
            break
        except Exception as e:
            last_err = str(e)
            time.sleep(1.5 * attempt)

    if last_err:
        df.at[i, "mistral_error"] = last_err

    # ì¤‘ê°„ ì €ì¥(20ê°œë§ˆë‹¤)
    if (i + 1) % 20 == 0:
        df.to_csv("/content/korea_policy_articles_summarized_partial.csv", index=False, encoding="utf-8-sig")

    time.sleep(SLEEP_BETWEEN_CALLS)


# 10) ì €ì¥ + ë‹¤ìš´ë¡œë“œ
df.to_csv(OUTPUT_PATH, index=False, encoding="utf-8-sig")
print("âœ… Saved:", OUTPUT_PATH)

files.download(OUTPUT_PATH)


```

```python
# ============================================================
# (Colab) Vectorize + FAISS + RAG(Mistral) end-to-end
# - CSV(ìš”ì•½+í‚¤ì›Œë“œ+ë³¸ë¬¸) -> ê²€ìƒ‰ìš© ë¬¸ì„œ êµ¬ì„± -> ì„ë² ë”© -> FAISS ì €ì¥
# - ì§ˆë¬¸ -> TopK ê²€ìƒ‰ -> Mistralë¡œ ê·¼ê±° ê¸°ë°˜ ë‹µë³€ ìƒì„±
# ============================================================

# ----------------------------
# 0) (í•„ìˆ˜) Mistral API Key ì„¤ì •
# ----------------------------
# âœ… ì•„ë˜ ì¤„ì˜ "sk-..."ë¥¼ ë³¸ì¸ í‚¤ë¡œ ë°”ê¾¸ê³  ì‹¤í–‰ë˜ê²Œ ë‘ì„¸ìš”.
#    (Secretsì— ì €ì¥í•´ë‘ì…¨ìœ¼ë©´ ì´ ì¤„ì€ ì£¼ì„ ì²˜ë¦¬í•´ë„ ë©ë‹ˆë‹¤.)
%env MISTRAL_API_KEY=Y2nJU0YqjbedsHpedWe76Y4PPGtllniO

# ----------------------------
# 1) ì„¤ì¹˜
# ----------------------------
!pip -q install pandas numpy sentence-transformers faiss-cpu tqdm mistralai

# ----------------------------
# 2) ì„í¬íŠ¸
# ----------------------------
import os
import re
import time
import numpy as np
import pandas as pd
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
import faiss
from mistralai import Mistral
from google.colab import files

# ----------------------------
# 3) ì…ë ¥ íŒŒì¼ ì°¾ê¸°(ìë™) + ì—†ìœ¼ë©´ ì—…ë¡œë“œ
# ----------------------------
CANDIDATES = [
    "/content/korea_policy_articles_summarized.csv",
    "/mnt/data/korea_policy_articles_summarized.csv",
    "korea_policy_articles_summarized.csv",
]

INPUT_PATH = next((p for p in CANDIDATES if os.path.exists(p)), None)
if INPUT_PATH is None:
    print("ğŸ“Œ ì…ë ¥ CSVë¥¼ ì°¾ì§€ ëª»í–ˆìŠµë‹ˆë‹¤. ì§€ê¸ˆ ì—…ë¡œë“œí•˜ì„¸ìš” (korea_policy_articles_summarized.csv).")
    up = files.upload()
    if not up:
        raise ValueError("âŒ ì—…ë¡œë“œëœ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
    fname = list(up.keys())[0]
    INPUT_PATH = f"/content/{fname}"

print("âœ… INPUT_PATH:", INPUT_PATH)

df = pd.read_csv(INPUT_PATH)

# ----------------------------
# 4) ì»¬ëŸ¼ ìë™ íƒì§€
# ----------------------------
def pick_col(cols, cands):
    for c in cands:
        if c in cols:
            return c
    return None

TITLE_COL = pick_col(df.columns, ["title", "ì œëª©"])
URL_COL   = pick_col(df.columns, ["url", "link", "URL"])
S1_COL    = pick_col(df.columns, ["summary_line1", "ìš”ì•½1", "summary1"])
S2_COL    = pick_col(df.columns, ["summary_line2", "ìš”ì•½2", "summary2"])
KW_COL    = pick_col(df.columns, ["keywords", "í‚¤ì›Œë“œ"])
BODY_COL  = pick_col(df.columns, ["content", "body", "text", "ë³¸ë¬¸", "ë‚´ìš©"])

if S1_COL is None or S2_COL is None:
    raise ValueError(f"âŒ ìš”ì•½ ì»¬ëŸ¼(summary_line1/2)ì„ ì°¾ì§€ ëª»í–ˆìŠµë‹ˆë‹¤. í˜„ì¬ ì»¬ëŸ¼: {list(df.columns)}")

print("âœ… Columns:", {
    "TITLE": TITLE_COL, "URL": URL_COL, "S1": S1_COL, "S2": S2_COL, "KW": KW_COL, "BODY": BODY_COL
})

def norm(x) -> str:
    if x is None:
        return ""
    try:
        if pd.isna(x):
            return ""
    except Exception:
        pass
    return re.sub(r"\s+", " ", str(x)).strip()

# ----------------------------
# 5) ê²€ìƒ‰ìš© ë¬¸ì„œ(ë²¡í„°í™” ëŒ€ìƒ) êµ¬ì„±
#    - ì§ˆë¬¸ ë§¤ì¹­ ê°•í™”ë¥¼ ìœ„í•´ "ì œëª© + ìš”ì•½2ì¤„ + í‚¤ì›Œë“œ + ë³¸ë¬¸ ë°œì·Œ"ë¥¼ ê²°í•©
# ----------------------------
MAX_BODY_CHARS = 2500

search_texts = []
meta_rows = []

for i, row in df.iterrows():
    title = norm(row[TITLE_COL]) if TITLE_COL else ""
    url   = norm(row[URL_COL]) if URL_COL else ""
    s1    = norm(row[S1_COL])
    s2    = norm(row[S2_COL])
    kw    = norm(row[KW_COL]) if KW_COL else ""
    body  = norm(row[BODY_COL]) if BODY_COL else ""

    if not (title or s1 or s2 or body):
        continue

    body_part = body[:MAX_BODY_CHARS] if body else ""

    parts = []
    if title: parts.append(f"ì œëª©: {title}")
    parts.append(f"ìš”ì•½: {s1} {s2}".strip())
    if kw: parts.append(f"í‚¤ì›Œë“œ: {kw}")
    if body_part: parts.append(f"ë³¸ë¬¸(ë°œì·Œ): {body_part}")

    doc = "\n".join(parts)
    search_texts.append(doc)

    meta_rows.append({
        "doc_id": len(meta_rows),
        "src_row": int(i),
        "title": title,
        "url": url,
        "summary_line1": s1,
        "summary_line2": s2,
        "keywords": kw,
    })

meta = pd.DataFrame(meta_rows)
print("âœ… docs:", len(search_texts))
if len(search_texts) == 0:
    raise ValueError("âŒ ë²¡í„°í™”í•  ë¬¸ì„œê°€ 0ê°œì…ë‹ˆë‹¤. ìš”ì•½/ë³¸ë¬¸ ì»¬ëŸ¼ì„ í™•ì¸í•˜ì„¸ìš”.")

# ----------------------------
# 6) ì„ë² ë”© ìƒì„± + FAISS ì¸ë±ìŠ¤ ìƒì„±/ì €ì¥
# ----------------------------
MODEL_NAME = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
model = SentenceTransformer(MODEL_NAME)

emb = model.encode(
    search_texts,
    batch_size=64,
    show_progress_bar=True,
    convert_to_numpy=True,
    normalize_embeddings=True,
).astype(np.float32)

print("âœ… embeddings:", emb.shape)

OUT_DIR = "/content"
os.makedirs(OUT_DIR, exist_ok=True)

meta_path = os.path.join(OUT_DIR, "embeddings_meta.csv")
faiss_path = os.path.join(OUT_DIR, "faiss.index")
model_path = os.path.join(OUT_DIR, "model_name.txt")

meta.to_csv(meta_path, index=False, encoding="utf-8-sig")
with open(model_path, "w", encoding="utf-8") as f:
    f.write(MODEL_NAME)

dim = emb.shape[1]
index = faiss.IndexFlatIP(dim)   # normalize_embeddings=Trueë©´ cosine ìœ ì‚¬ë„
index.add(emb)
faiss.write_index(index, faiss_path)

print("âœ… Saved:", faiss_path, meta_path, model_path)

# ----------------------------
# 7) TopK ê²€ìƒ‰ í•¨ìˆ˜
# ----------------------------
def search_topk(query: str, topk: int = 5):
    q_emb = model.encode([query], normalize_embeddings=True, convert_to_numpy=True).astype(np.float32)
    D, I = index.search(q_emb, topk)
    hits = []
    for score, idx in zip(D[0].tolist(), I[0].tolist()):
        if idx < 0:
            continue
        h = meta.iloc[idx].to_dict()
        h["score"] = float(score)
        h["doc_text"] = search_texts[idx]
        hits.append(h)
    return hits

# ----------------------------
# 8) Mistral RAG ë‹µë³€ ìƒì„± í•¨ìˆ˜
# ----------------------------
api_key = os.getenv("MISTRAL_API_KEY", "Y2nJU0YqjbedsHpedWe76Y4PPGtllniO").strip()
if not api_key:
    raise ValueError("âŒ MISTRAL_API_KEYê°€ ì—†ìŠµë‹ˆë‹¤. ìœ„ %env ì¤„ì„ ì„¤ì •í•˜ê±°ë‚˜ Colab Secretsì— ì €ì¥í•˜ì„¸ìš”.")

mistral_client = Mistral(api_key=api_key)
MISTRAL_MODEL = "ministral-8b-latest"

def rag_answer_with_mistral(question: str, hits: list, max_chars_per_doc: int = 1800) -> str:
    ctx = []
    for j, h in enumerate(hits, start=1):
        txt = h["doc_text"][:max_chars_per_doc]
        ctx.append(f"[ê·¼ê±° {j}] score={h['score']:.3f}\n{txt}\n")

    system = """ë„ˆëŠ” ì •ì±…ë‰´ìŠ¤(ìš”ì•½/í‚¤ì›Œë“œ/ë³¸ë¬¸ ë°œì·Œ) ê·¼ê±°ë¡œë§Œ ë‹µí•˜ëŠ” ì •ì±… ë¶„ì„ê°€ë‹¤.
ê·œì¹™:
- ì•„ë˜ ì œê³µëœ 'ê·¼ê±°' ì•ˆì—ì„œ í™•ì¸ ê°€ëŠ¥í•œ ë‚´ìš©ë§Œìœ¼ë¡œ ë‹µí•´ë¼.
- ê·¼ê±°ì— ì—†ëŠ” ì‚¬ì‹¤/ìˆ˜ì¹˜/ì œë„ ë³€ê²½ì„ ì ˆëŒ€ ì§€ì–´ë‚´ì§€ ë§ˆë¼.
- ë‹µë³€ ë§ˆì§€ë§‰ì— ì‚¬ìš©í•œ ê·¼ê±° ë²ˆí˜¸ë¥¼ [ê·¼ê±° 1]ì²˜ëŸ¼ ëª…ì‹œí•´ë¼.
- í•œêµ­ì–´ë¡œ ì‘ì„±.
"""

    user = f"""ì§ˆë¬¸: {question}

ì•„ë˜ ê·¼ê±°ë“¤ì„ ë°”íƒ•ìœ¼ë¡œ ë‹µí•´ì¤˜.
{chr(10).join(ctx)}
"""

    resp = mistral_client.chat.complete(
        model=MISTRAL_MODEL,
        messages=[
            {"role": "system", "content": system},
            {"role": "user", "content": user},
        ],
        temperature=0.2,
        max_tokens=900,
    )
    return resp.choices[0].message.content

# ----------------------------
# 9) ì‹¤í–‰ ì˜ˆì‹œ: "R&Dì‚¬ì—…" ì§ˆë¬¸
# ----------------------------
QUESTION = "R&Dì‚¬ì—… ê´€ë ¨í•´ì„œ ìµœê·¼ ì •ì±… ë‚´ìš©ê³¼ ì§€ì› ë°©í–¥ì„ ìš”ì•½í•´ì¤˜."

hits = search_topk(QUESTION, topk=5)

print("\n=== TOP HITS ===")
for h in hits:
    print(f"- score={h['score']:.3f} | {h['title']} | {h['url']}")

answer = rag_answer_with_mistral(QUESTION, hits)
print("\n=== RAG ANSWER ===\n")
print(answer)

# ----------------------------
# 10) (ì„ íƒ) íŒŒì¼ ë‹¤ìš´ë¡œë“œ
# ----------------------------
files.download("/content/faiss.index")
files.download("/content/embeddings_meta.csv")
files.download("/content/model_name.txt")
```
